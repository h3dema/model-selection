```python
import numpy as np
import matplotlib.pyplot as plt

from gilbert_model import GilbertModel
from gilbert_model import compute_transition_probabilities
```

```python
N = 10000
model = GilbertModel()

samples = model.sample(N)
print("First 20 samples:", samples[:20])
```

```python
plt.figure(figsize=(12, 4))
# I find it more intuitive to plot CORRUPT as 0 and CORRECT as 1
plt.step(range(N), 1 - samples, where='post')

plt.title("Gilbert Model State Over Time")
plt.xlabel("Time")
plt.ylabel("State")
plt.yticks([0, 1], ["CORRUPT", "CORRECT"])

plt.grid(True)
plt.tight_layout()
plt.savefig("gilbert_model.pdf")
plt.show()
```

## compute transition probabilities

```python
probs = compute_transition_probabilities(samples)
print("\nSampled Transition Probabilities:")
for (i, j), prob in probs.items():
    print(f"Pr({'c' if j == 0 else 'p'} | {'c' if i == 0 else 'p'}) = {prob:.4f}")
```
