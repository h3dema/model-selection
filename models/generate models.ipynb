```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import numpy as np

from models import generate_models
```

```python
M = 10  # number of models to generate

models = generate_models(M, C=[1, 2, 3, 4, 5])
print(f"Generated {len(models)} models.")
```

```python

fig, ax = plt.subplots(2, 1, figsize=(10, 15))

# energy hist
ax[0].hist([model.energy for model in models], bins=10)
ax[0].set_title('Energy Distribution')
ax[0].set_xlabel('Energy')
ax[0].set_ylabel('Count')

# Bin sizes and count how many models fall into each bin
sizes = [model.size for model in models]
size_bins = 10  # You can adjust this
counts, bins = np.histogram(sizes, bins=size_bins)

# Compute bin centers for plotting
bin_centers = 0.5 * (bins[1:] + bins[:-1])

# Plot size distribution
ax[1].bar(bin_centers, counts, width=(bins[1] - bins[0]) * 0.9)
ax[1].set_xlabel('Model Size')
ax[1].set_ylabel('Number of Models')
ax[1].set_title('Model Count per Size Range')
ax[1].yaxis.set_major_locator(MaxNLocator(integer=True))

for axis in ax:
    axis.yaxis.set_major_locator(MaxNLocator(integer=True))

plt.tight_layout()
plt.show()
```
