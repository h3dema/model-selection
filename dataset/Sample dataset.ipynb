```python
from collections import Counter
import matplotlib.pyplot as plt

from dataset import create_classes, create_dataset, create_subsets, assign_classes
```

```python
def plot_class_distribution(subsets, class_D):
    for name, subset in subsets.items():
        # Count classes in the current subset
        class_counts = Counter([class_D[e] for e in subset])
        labels = [f"Class {cls}" for cls in class_counts.keys()]
        sizes = list(class_counts.values())

        # Plot pie chart
        plt.figure(figsize=(5, 5))
        plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
        plt.title(f"Class Distribution in {name}")
        plt.axis('equal')  # Equal aspect ratio ensures pie is circular
        plt.show()
```

```python
# Example configuration
num_classes = 5
size_dataset = 2000

subset_sizes = [13, 20, 17]
num_subsets = 3
```

```python
# Create C and D
C = create_classes(num_classes)
D = create_dataset(size_dataset)
```

```python
assert num_subsets == len(subset_sizes)
```

```python
class_D = assign_classes(D, C)
subsets = create_subsets(D, subset_sizes)

# Display results
print("Class set C:", C)
# print("\nDataset D with assigned classes:")
# for e in D:
#     print(f"{e}: class {class_D[e]}")

print("\nSubsets D_i:", len(subsets))
# for name, subset in subsets.items():
#     print(f"{name}: {subset}")
```
